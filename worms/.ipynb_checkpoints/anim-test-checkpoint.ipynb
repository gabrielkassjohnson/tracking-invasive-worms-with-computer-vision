```python
# Import Libraries 
import matplotlib
import matplotlib.pyplot as plt
import matplotlib.animation as animation
import matplotlib.image


import cv2
import os
import PIL
from PIL import Image 
from glob import glob
import numpy as np
%matplotlib notebook
```

```python
dir = os.path.join("data", "*.jpg")
file_list = [file for file in glob(dir)]
print(file_list)
```

```python
file_list[0]
```

```python

# Create the frames
img_array = []
for file in file_list:
    img = Image.open(file)
    img_array.append(img)
    print(file)

# Save into a GIF file that loops forever
img_array[0].save('test.gif', format='GIF',
               append_images=img_array[1:],
               save_all=True,
               duration=180, loop=0)
```

```python
img_array = []
for filename in file_list:
    img = matplotlib.image.imread(filename)
    imgplot = plt.imshow(img)

    img_array.append([imgplot])
```

```python
img_array
```

```python

fig = plt.figure()
ax = plt.gca()


## create an instance of animation
anim = animation.ArtistAnimation(fig, img_array, interval=1000, blit=True, repeat_delay=1000)

## NB: The 'save' method here belongs to the object you created above
anim.save("anim.gif")

## Showtime!
plt.show()
```

```python
import cv2
import numpy as np
import os

image_folder = 'data'
video_file = 'vid.avi'
image_size = (1600, 1200)
fps = 1

images = [img for img in os.listdir(image_folder) if img.endswith(".jpg")]
images.sort()
file_list.sort()
out = cv2.VideoWriter(video_file, cv2.VideoWriter_fourcc(*'MJPG'), fps, image_size)

img_array = []
for filename in file_list:
    img = cv2.imread(filename)
    img_array.append(img)
    out.write(img)

out.release()
```

```python
img_list = []
cv2_img_list= []
im_list = []
for file in file_list:
    img = Image.open(file)
    img_list.append(img)
    cv2_img = cv2.imread(file)
    cv2_img_list.append(cv2_img)
    imgplot = plt.imshow(img,animated=True)
    im_list.append([imgplot])
print(im_list)
```

```python
im_list[0]
```

```python

fig= plt.figure(dpi=200)

ani = animation.ArtistAnimation(fig, im_list)
writer = animation.FFMpegWriter(fps=2)
ani.save('file.gif',writer=writer)

plt.show()

```

```python
from matplotlib import animation  
import matplotlib.image as mgimg
import numpy as np

#set up the figure
fig = plt.figure()
ax = plt.gca()

#initialization of animation, plot array of zeros 
def init():
    imobj.set_data(np.zeros((100, 100)))

    return  imobj,

def animate(i):
    ## Read in picture
    fname = img_array[i] 

    ## here I use [-1::-1], to invert the array
    # IOtherwise it plots up-side down
    img = mgimg.imread(fname)[-1::-1]
    imobj.set_data(img)

    return  imobj,


## create an AxesImage object
imobj = ax.imshow( np.zeros((100, 100)), origin='lower', alpha=1.0, zorder=1, aspect=1 )


anim = animation.FuncAnimation(fig, animate, init_func=init, repeat = True,
                               frames=range(1,4), interval=200, blit=True, repeat_delay=1000)

anim.save()
plt.show()
```
